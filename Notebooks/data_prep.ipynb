```
%r

dataset <- data.frame()

# Add data from each file in turn
for (k in seq_along(data_files)) {
  
  cat("Reading file", k, "/", length(data_files), " and file: ", data_files[k], "\n")
  if(data_files[k] != "/dbfs/data/CSV//Data Overview.csv"){
    
     d <- read_csv(data_files[k])
  
    # Add a column with a label representing the person
    cat("\n", as.numeric(paste0(unlist(stringr::str_extract_all( data_files[k], "[[:digit:]]")),collapse='')), " file: ",data_files[k], "\n")    
    d$trip <- as.numeric(paste0(unlist(stringr::str_extract_all( data_files[k], "[[:digit:]]")),collapse=''))

    dataset <- bind_rows(dataset, d)
  }
 
}

head(dataset)
```

```
%r
data_summary <- read_csv("/dbfs/data/CSV//Data Overview.csv", skip = 1)
data_summary <- data_summary %>% 
  select(`Date Collected`, `File Name`, `Taxi Route`, `Driving Style`) %>%
  mutate(trip = as.numeric(substr(data_summary$`File Name`, 1, 2)))

data_summary %>% tail()
```

```
%r
df <- df %>% 
  filter(`Time (s)` <= 1222)

df %>% 
  head()
```

```
%r
df_acc <- df %>%
  select(`Time (s)`, "X_Accel", "Y_Accel", "Z_Accel", "Driving Style", "trip") %>%
  transform(X_Accel = as.numeric(X_Accel), 
            Y_Accel = as.numeric(Y_Accel),
            Z_Accel = as.numeric(Z_Accel),
            `Driving Style` = as.numeric( as.factor(`Driving Style`) ))

as.matrix(df_acc[1:1222, -1])
```

```
%r
df_acc <- df %>%
  select(`Time (s)`, "X_Accel", "Y_Accel", "Z_Accel", "Driving Style", "trip") %>%
  transform(X_Accel = as.numeric(X_Accel), 
            Y_Accel = as.numeric(Y_Accel),
            Z_Accel = as.numeric(Z_Accel),
            `Driving Style` = as.numeric( as.factor(`Driving Style`) ))
# Reshape data into 3 dimensions:
#   0-dimension ("rows") is observations (178348 in total) and 72 trips
#   1-dimension ("columns") is time series values (1222)
#   2-dimension ("leaves") are as follows (5 in total):
#     * 3 directions (x-, y-, z-acceleration)
#     * Activity type labels
#     * Trip labels

reshaped_data <- array(0, dim = c(72, 1222, 5))

for (k in seq_len(72)) {
  
  start <- 1222*(k-1) + 1
  stop <- start + 1221
  
  # Else copy all but count column to the new data block
  reshaped_data[k, , ] <- as.matrix(df_acc[start:stop, -1])
}

reshaped_data
```

```
%r

driving_x <- reshaped_data[,,1:3] %>%
  apply(c(1, 3), scale) %>% # scaling within each series so no issue with train/test
  aperm(c(2,1,3))           # undo apply's ridiculous transpose

dimnames(driving_x) <- list(NULL, NULL, c("X_Accel", "Y_Accel", "Z_Accel"))

driving_labels <- reshaped_data[,1,5] %>%
  as.integer()

drivingData <- list(x = driving_x, 
                    labels = driving_labels)

saveRDS(drivingData, "/dbfs/data/drivingData.rds")
saveRDS(driving_x, "/dbfs/data/driving_x.rds")
saveRDS(driving_labels, "/dbfs/data/driving_labels.rds")
```
